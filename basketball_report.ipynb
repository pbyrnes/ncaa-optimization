# Introduction

Each year the NCAA men's basketball tournament involves teams traveling throughout the United States to play games.  Current practice is to have a human committee determine which teams should play where through an unspecified process.  One, perhaps secondary, goal is to reduce the total amount of travel of all teams.  This has benefits of reduced travel costs and potential for more direct fan interaction (since fans are more likely to go to games closer to their home than games very far away).  The goal of this project is to create an optimal bracket that maintains seeds while reducing total travel distance.


# Problem Description

The problem starts with each of 64 teams having data that include the latitude and longitude of their home gym along with a seed (4 teams with each possible seed from 1-16).  There also are 8 first round sites that have information consisting of their latitude and longitude.  Each team is assigned to a site in such a way that each site gets 2 complete pods.  A pod consisting of a group of teams whose seeds are one of the sets $\{1, 16, 8, 9\}$, $\{2, 15, 7, 10\}$, $\{3, 14, 6, 11\}$, and $\{4, 13, 5, 12\}$.  We will call these pod types 1, 2, 3, and 4 respectively.  The goal is to minimize total travel distance for the 64 teams during the first-round in the tournament.

The formal mathematical definition involves parameters:

* $D_{i,k}$ for $1 \leq i \leq 64$, $1 \leq k \leq 8$: the distance in miles between team $i$'s home gym amd site $k$.

The variables are:

* $x_{a,k} \in \{0,1,2\}$ for $1 \leq a \leq 4$, $1 \leq k \leq 8$: the number of pods of type $a$ assigned to site $k$.
* $z_{i,k} \in \{0, 1\}$ for $1 \leq i \leq 64$, $1 \leq k \leq 8$: binary variable that is 1 exactly when team $i$ is assigned to site $k$.

The constraints are:

1) $$\sum_{k=1}^8 z_{i,k} = 1$$ for $1 \leq i \leq 64$  (each team plays at exactly one site)
2) $$\sum_{a=1}^4 x_{a, k} = 2$$ for $1 \leq k \leq 8$  (each site has exactly 2 pods)
3) $$\sum_{i \text{ such that school } i \text{ has seed } s} z_{i, k} = x_{a, k}$$ for each $1 \leq a \leq 4$, $s$ in pod $a$, and $1 \leq k \leq 8$ (each site gets the correct number of teams for its assigned pods)

The objective is to minimize the sum of the distances which is:
$$ \sum_{i=1}^{64} \sum_{k=1}^8 D_{i,k}\cdot z_{i,k}$$

# Data Collection

For the purposes of this project, distances will be computed as geodesics from home arena locations to the tournament playing site locations.  As of 2026, the NCAA men's basketball tournament starts with 68 teams, but 4 of those teams are eliminated in "play-in" games before the rest of the tournament begins.  We will only consider the 64 teams not eliminated in the play-in games.

To gather data, we will use the 2026 NCAA men's tournament with information available at [https://en.wikipedia.org/wiki/2026_NCAA_Division_I_men's_basketball_tournament](https://en.wikipedia.org/wiki/2026_NCAA_Division_I_men's_basketball_tournament).  Home arena coordinates for each school can be found at [https://en.wikipedia.org/wiki/List_of_NCAA_Division_I_basketball_arenas](https://en.wikipedia.org/wiki/List_of_NCAA_Division_I_basketball_arenas).  Site coordinates are found by clicking on the links for each first-round site at the 2026 tournament page.

This project uses the geopy library for computing geodesic distances.  Data files are provided in the project repository.

# Method

The integer program was solved using gurobipy.  Runtime is less than a second.

# Results

The original tournament for 2026 had a total first-round travel distance of $54,495$ miles.  The travel-optimized tournament had a total first-round travel distance of $35,984$ miles.  So, minimizing travel distance resulted in a tournament with roughly 66% of the original travel for first-round games.

# Conclusions

This project shows that it was possible to assign NCAA men's basketball teams to first-round March Madness locations such that total travel distance would be reduced by roughly one-third while keeping all the seeds the same.  A tournament with more teams playing closer to home has potential to increase engagement due to easier fan travel and increased chances of regional rivals interacting with each other.

It should be noted that there are further constraints to team assignments (involving certain teams not playing at certain locations or facing certain other teams) that may make the found solution infeasible.  Other considerations not accounted for in this project include giving some amount of location advantage to higher-seeded teams and the relative ranking of the four teams with a given seed.
