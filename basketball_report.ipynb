# Introduction

Each year the NCAA men's basketball tournament involves teams traveling throughout the United States to play games.  Current practice is to have a human committee determine which teams should play where through an unspecified process.  One, perhaps secondary, goal is to reduce the total amount of travel of all teams.  This has benefits of reduced travel costs and potential for more direct fan interaction (since fans are more likely to go to games closer to their home than games very far away).  The goal of this project is to create an optimal bracket that maintains seeds while reducing total travel distance.


# Problem Description

The goal is to minimize total travel distance for the 64 teams during the first-round in the tournament.  There are some constraints on creating the bracket.  There are eight sites that are chosen for each year.  Each site hosts two pods.  Each pod consists of four teams with one of four possible sets of seeds.   

For the purposes of this project, distances will be computed as geodesics from home arena locations to the tournament playing site locations.  As of 2026, the NCAA men's basketball tournament starts with 68 teams, but 4 of those teams are eliminated in "play-in" games before the rest of the tournament begins.  We will only consider the 64 teams not eliminated in the play-in games.

The tournament structure starts with seeding each of the 64 teams with a seed from 1 to 16.  This results in four teams with each possible seed.  Further, teams are grouped into pods that include one team each of a certain set of seeds.  The possible pods are seeds $\{1, 16, 8, 9\}$, $\{2, 15, 7, 10\}$, $\{3, 14, 6, 11\}$, and $\{4, 13, 5, 12\}$.  Each of the 8 first-round playing sites have 2 pods (possible pods with the same seeds).

This project is to investigate possible tournaments that keep the pod structure the same and the team seeds the same.  The goal is to find the potential tournament that minimizes the total travel distance for all teams.  There are reasons this may not be an ideal tournament.  First off, there are other rules about the bracket such as certain teams not playing on Sunday and certain team matchups being disallowed.  But, this distance-optimal bracket at least gives us a lower bound on total travel distance.

To gather data, we will use the 2026 NCAA men's tournament with information available at [https://en.wikipedia.org/wiki/2026_NCAA_Division_I_men's_basketball_tournament](https://en.wikipedia.org/wiki/2026_NCAA_Division_I_men's_basketball_tournament).  Home arena coordinates for each school can be found at [https://en.wikipedia.org/wiki/List_of_NCAA_Division_I_basketball_arenas](https://en.wikipedia.org/wiki/List_of_NCAA_Division_I_basketball_arenas).  This project uses the geopy library for computing geodesic distances.
